```python
#pork-only
NC=[]
for i in range(len(Tech)):
    if Time[i] <2025:
        NC.append(OC[i])

    else:
        has_keyword = False
        for keyword in SubsidyT:
            if keyword in Tech[i]:
                NC.append(OC[i] * 0.6) 
                has_keyword = True
                break
        if not has_keyword:
            NC.append(OC[i])
print(len(NC))
```

```python
#pork-and-stick, wind and solar

NC=[]
for i in range(len(Tech)):
    int(Time[i])
    if Time[i] < 2025:
        NC.append(OC[i])
        
    elif Time[i] >2035:
        NC.append(OC[i]*0.87)

    else:
        has_keyword = False
        for keyword in SubsidyT:
            if keyword in Tech[i]:
                NC.append(OC[i] * 0.6) 
                has_keyword = True
                break
        if not has_keyword:
            NC.append(OC[i])
print(len(NC))
```

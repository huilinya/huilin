```python
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt
import scienceplots as sp
```

```python
SecEm=pd.read_excel('emissions data.xlsx', sheet_name="pork")
Sector=SecEm['Sector']
Em1990=SecEm["em1990"]
Em2005=SecEm["em2005"]
Em2010=SecEm["em2010"]
Em2015=SecEm["em2015"]
Em2020=SecEm["em2020"]
Em2025=SecEm["em2025"]
Em2030=SecEm["em2030"]
Em2035=SecEm["em2035"]
Em2040=SecEm["em2040"]
Em2045=SecEm["em2045"]
Em2050=SecEm["em2050"]
E1990=sum(Em1990)*44/12
E2005=sum(Em2005)*44/12
E2010=sum(Em2010)*44/12
E2015=sum(Em2015)*44/12
E2020=sum(Em2020)*44/12
E2025=sum(Em2025)*44/12
E2030=sum(Em2030)*44/12
E2035=sum(Em2035)*44/12
E2040=sum(Em2040)*44/12
E2045=sum(Em2045)*44/12
E2050=sum(Em2050)*44/12

print(
    "2005 emission", E2005, "MtCO2",
    "2050 emission", E2050, "MtCO2",
    "Difference",(E2005-E2050)/E2005
)
```

```python
SecEmB=pd.read_excel('emissions data.xlsx', sheet_name="baseline")#B=Baseline
SectorB=SecEmB['Sector']
Em1990B=SecEmB["em1990"]
Em2005B=SecEmB["em2005"]
Em2010B=SecEmB["em2010"]
Em2015B=SecEmB["em2015"]
Em2020B=SecEmB["em2020"]
Em2025B=SecEmB["em2025"]
Em2030B=SecEmB["em2030"]
Em2035B=SecEmB["em2035"]
Em2040B=SecEmB["em2040"]
Em2045B=SecEmB["em2045"]
Em2050B=SecEmB["em2050"]
EB_1990=sum(Em1990B)*44/12
EB_2005=sum(Em2005B)*44/12
EB_2010=sum(Em2010B)*44/12
EB_2015=sum(Em2015B)*44/12
EB_2020=sum(Em2020B)*44/12
EB_2025=sum(Em2025B)*44/12
EB_2030=sum(Em2030B)*44/12
EB_2035=sum(Em2035B)*44/12
EB_2040=sum(Em2040B)*44/12
EB_2045=sum(Em2045B)*44/12
EB_2050=sum(Em2050B)*44/12

print(
    "2005 emission", EB_2005, "MtCO2",
    "2050 emission", EB_2050, "MtCO2",
    "Difference",(EB_2005-EB_2050)/EB_2005
)
```

```python
SecEmPaS=pd.read_excel('emissions data.xlsx', sheet_name="pork and stick")#PaS=pork and stick
SectorPaS=SecEmPaS['Sector']
Em1990PaS=SecEmPaS["em1990"]
Em2005PaS=SecEmPaS["em2005"]
Em2010PaS=SecEmPaS["em2010"]
Em2015PaS=SecEmPaS["em2015"]
Em2020PaS=SecEmPaS["em2020"]
Em2025PaS=SecEmPaS["em2025"]
Em2030PaS=SecEmPaS["em2030"]
Em2035PaS=SecEmPaS["em2035"]
Em2040PaS=SecEmPaS["em2040"]
Em2045PaS=SecEmPaS["em2045"]
Em2050PaS=SecEmPaS["em2050"]
EPS_1990=sum(Em1990PaS)*44/12
EPS_2005=sum(Em2005PaS)*44/12
EPS_2010=sum(Em2010PaS)*44/12
EPS_2015=sum(Em2015PaS)*44/12
EPS_2020=sum(Em2020PaS)*44/12
EPS_2025=sum(Em2025PaS)*44/12
EPS_2030=sum(Em2030PaS)*44/12
EPS_2035=sum(Em2035PaS)*44/12
EPS_2040=sum(Em2040PaS)*44/12
EPS_2045=sum(Em2045PaS)*44/12
EPS_2050=sum(Em2050PaS)*44/12

Deviation=(EPS_2025+EPS_2030+EPS_2035+EPS_2040+EPS_2045+EPS_2050-(EB_2025+EB_2030+EB_2035+EB_2040+EB_2045+EB_2050))/(EB_2025+EB_2030+EB_2035+EB_2040+EB_2045+EB_2050)

print(
    "2005 emission", EPS_2005, "MtCO2",
    "2050 emission", EPS_2050, "MtCO2",
    "Difference",(EPS_2005-EPS_2050)/EPS_2005,
    "Deviation",Deviation
)
```

```python
Year=["2005","2010","2015","2020","2025","2030","2035","2040","2045","2050"]
Emission=[E2005,E2010,E2015,E2020,E2025,E2030,E2035,E2040,E2045,E2050]
EmissionB=[EB_2005,EB_2010,EB_2015,EB_2020,EB_2025,EB_2030,EB_2035,EB_2040,EB_2045,EB_2050]
# EmissionPO=[EPO_2005,EPO_2010,EPO_2015,EPO_2020,EPO_2025,EPO_2030,EPO_2035,EPO_2040,EPO_2045,EPO_2050]
EmissionPaS=[EPS_2005,EPS_2010,EPS_2015,EPS_2020,EPS_2025,EPS_2030,EPS_2035,EPS_2040,EPS_2045,EPS_2050]
legend=["Baseline","Pork (stick)","Pork-and-stick"]

fig, ax = plt.subplots(1, figsize=(12, 7))
# plt.size((20,10))
plt.style.use(['science', 'no-latex'])
plt.title('Emission trajectory, MtCO2')

plt.plot(Year,EmissionB,color="blue")
plt.plot(Year,Emission,color="tomato")
plt.plot(Year,EmissionPaS,color="green")

plt.legend(legend)
Emission_difference={
    "Year":Year,
    "Baseline Emissions":EmissionB,
    "Pork Emissions": Emission,
    "Pork-and-stick emissions": EmissionPaS
}
ED=pd.DataFrame(Emission_difference)
ED.to_csv("Emission trajectories.csv")
```
